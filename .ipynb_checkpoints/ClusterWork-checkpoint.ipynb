```python
def alejogfactor(r,M,a,lam):
    return np.sqrt(r ** 3 - 3 * M*r**2 + 2 * a * np.sqrt(M) * r ** (3/2))/(r ** (3/2) + np.sqrt(M)*(a - lam))
    
def swargfactor(r,mass):
    return np.sqrt(1-2*ilp.G*mass/(ilp.c ** 2 * r))
```

```python
alejoG = alejogfactor(rarrayRG,1,spin_case,0)
```

```python
def rpeak(r,array):
    return r[np.argmax(array)]
```

```python

for i in range(3):
    r =bigRun["ModelC"+ str(i+1) + "1"]["np"]["radVnu_thick"]() 
    DataR230_1[i] = float(r[np.abs((xvar-230e9)).argmin(),3])
    
    r =bigRun["ModelC"+ str(i+1) + "2"]["np"]["radVnu_thick"]() 
    DataR230_2[i] = float(r[np.abs((xvar-230e9)).argmin(),3])
    
    
    r =bigRun["ModelC"+ str(i+1) + "3"]["np"]["radVnu_thick"]() 
    DataR230_3[i] = float(r[np.abs((xvar-230e9)).argmin(),3])
    
    
    r =bigRun["ModelC"+ str(i+1) + "1"]["np"]["radVnu_thick"]() 
    TDataR230_1[i] = float(r[np.abs((xvar-10e9)).argmin(),3])
    
    r =bigRun["ModelC"+ str(i+1) + "2"]["np"]["radVnu_thick"]() 
    TDataR230_2[i] = float(r[np.abs((xvar-10e9)).argmin(),3])
    
    
    r =bigRun["ModelC"+ str(i+1) + "3"]["np"]["radVnu_thick"]() 
    TDataR230_3[i] = float(r[np.abs((xvar-10e9)).argmin(),3])
```

```python
bpUnit['p_temp']
```

```python
rg = ilp.rg_func(bpUnit['mass'])
rarrayRG = (np.linspace(2,30,300))
rarray = (np.linspace(2,30,300) * rg).value
rarrayUnit = np.linspace(2,30,300) * rg 
te = ilp.te_func(rarray,bpUnit['mass'],bpUnit['rb_0'],bpUnit['t_e0'],bpUnit['p_temp'])
sgfactor = swargfactor(rarrayUnit,bpUnit['mass'])
theta_e = ilp.theta_e_func(te)

gR = np.array([1.0,1,.0,1.0])

for i in range(3):
    te = ilp.te_func(rarray,bpUnit['mass'],bpUnit['rb_0'],bpUnit['t_e0'],-alphaT[i])
    sgfactor = swargfactor(rarrayUnit,bpUnit['mass'])
    gR[i] = rpeak(rarrayRG,(sgfactor * te))
```
